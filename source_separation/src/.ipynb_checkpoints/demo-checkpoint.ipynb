# Convex Analysis of Mixture Non-negative Sources (CAMNS)

The method implemented below solves the problem of non-negative blind source separation. The general idea is quite simple, although it contains many assumptions.

Assumptions:

(A1) $\textbf{Source non-negativity}$: For each $j, s_j \in \mathbb{R}_{+}^{L}$.\
(A2) $\textbf{Local dominance}$: For each $i \in {1, \ldots, N}$, there exists an (unknown) index $l_i$ such that $s_i[l_i] > 0$ and $s_i[l_i] = 0, \forall j \ne i$.\
(A3)$\textbf{Unit row sum:}$ For all $i = 1, \ldots, M,\ \sum_{j=1}^{N}a_{ij}=1.$\
(A4)$M \geq N$ and $A$ is of full column rand.

Under these assumptions we can find the source in this way:

$$p^* = \min_{\alpha:C\alpha+d\geq0}r^T(C\alpha + d)$$
$$q^* = \max_{\alpha:C\alpha+d\geq0}r^T(C\alpha + d)$$

Let's demonstrate the work of the algorithm:

```python
from image import camns_image, plot_images
from audio import camns_audio
```

Load the testing images:

```python
img1 = camns_image("cao1.jpg")
img2 = camns_image("ksiwek1.jpg")
img3 = camns_image("zhang1.jpg")
init = [img1, img2, img3]
```

```python
plot_images(init, 1, 3)
```

```python
mixed = list(camns_image.mix(init))
for i, img in enumerate(mixed):
    img.write("img_mixed_{}.jpg".format(i))
```

Let's see the mixed images:

```python
plot_images(mixed, 1, 3)
```

```python
unmixed = list(camns_image.camns_lp(mixed))
```

```python
plot_images(unmixed, 1, 3)
```

```python
plot_images(init + mixed + unmixed, 3, 3)
```

Thus as we can see, we received really good results!

Let's check the whole process together on another example.

```python
img_cat = camns_image("cat.jpg")
img_helic = camns_image("helicopters.jpg")
init = [img_cat, img_helic]
mixed = list(camns_image.mix(init))
for i, img in enumerate(mixed):
    img.write("img_mixed_mem_{}.jpg".format(i))
unmixed = list(camns_image.camns_lp(mixed))
plot_images(init + mixed + unmixed, 3, 2)
```

So we have quite a powerful tool not only for separating sources but also for creating memes and extracting pictures from memes!

# Audio signals separating

This algorithm works in the same way for any type of signal, regatdless its origin. It operates with vector representation of signal. So we can try to use it with audio, for instance!

```python
from audio import camns_audio
```

```python
rain = camns_audio("rain.wav")
bob = camns_audio("bob.wav")
city = camns_audio("city.wav")
init = [rain, bob, city]

mixed = list(camns_audio.mix([rain, bob, city]))
for i, sound in enumerate(mixed):
    sound.write("mixed_{}.wav".format(i))

unmixed = list(camns_audio.camns_lp(mixed))
for sound, i in enumerate(mixed):
    sound.write("unmixed_{}.wav".format(i))
```
